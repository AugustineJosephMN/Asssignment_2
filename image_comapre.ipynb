```
img_y,  img_x  = img_arr.shape[:2]
sub_y, sub_x = sub_arr.shape[:2]
```

```
print(img_y,  img_x )
print(sub_y, sub_x)
```

```
limit_x = img_x - sub_x + 1
limit_y = img_y - sub_y + 1
```

```
for x in range(0, limit_x):
    for y in range(0, limit_y):
        x2 = x + sub_x
        y2 = y + sub_y
        pic = img_arr[y:y2, x:x2]
        test = (pic == sub_arr)
        if test.all():
            print(x,y)
            break
```
